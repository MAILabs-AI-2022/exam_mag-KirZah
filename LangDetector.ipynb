```python
def get_words_from_file(filename: str) -> list:
    with open(filename) as f:
        lines = f.readlines()
    for i, line in enumerate(lines):
        lines[i] = lines[i][:-1]
    return lines

def get_max_word_length(words: list) -> int:
    max_len = 0
    for i, line in enumerate(lines):
        cur_len = len(lines[i])
        if max_len < cur_len:
            max_len = cur_len
    
    return max_len


dutch_words = get_words_from_file('dutch.txt')
hungarian_words = get_words_from_file('hungarian.txt')
portugese_words = get_words_from_file('portugese.txt')
print(get_max_word_length(dutch_words))
print(get_max_word_length(hungarian_words))
print(get_max_word_length(portugese_words))
```

```python
def make_dataframe(words, class_id, starting_index):
    lst2 = [class_id for _ in words]
    print(len(lst2))
    df = pd.DataFrame(
        list(zip(words, lst2)),
        columns=['word', 'class_id'],
        index=[starting_index + i for i in range(len(words))]
    )
    return df

df_dutch = make_dataframe(dutch_words, 0, starting_index=0)
df_hungarian = make_dataframe(hungarian_words, 1, starting_index=len(df_dutch))
df_portugese = make_dataframe(portugese_words, 2, starting_index=len(df_dutch)+len(df_hungarian))

df_dutch.head()
# df_hungarian.head()
# df_portugese.head()

df = pd.concat([df_dutch, df_hungarian, df_portugese])
df

# todo: описать модель
# todo: предаточная функция, функция ошибки, кодировка слов.

# кодировка слов:
# простой базовый классификатор... слова будут кодироваться в ANSII формат
# слова лучше токенизировать
# после padding всех слов до длины в 20 символов слова будут векторизованы

```
